### 2.1 Traffic throughout the day
- Let's look at a breakdown of when the site receives it's traffic

```python
# Add an hour column from our datetime index
df["hour"] = df.index.hour
# set a datetime column aside from our index
df["datetime"] = df.index

# set a weekday column so we can split weekday and weekend traffic
df["weekday"] = df.index.weekday

# Mon-Fri=0-4; Sat-Sun=5-6
df_weekday = df[df.index.weekday < 5].copy()
df_weekend = df[df.index.weekday >= 5].copy()
```

```python
# let's create some plots that show the traffic breakdown
import matplotlib.patches as mpatches

# create some color patches and labels
average_day_leg = mpatches.Patch(color="g", label="Average Day")
weekday_leg = mpatches.Patch(color='b', label='Weekday')
weekend_leg = mpatches.Patch(color='r', label='Weekend')

# create subplots of correct size
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(9,10))
fig.tight_layout()
plt.subplots_adjust(hspace = 0.3) # adjust the space between the subplots

# plot total users per country
df.hour.value_counts().sort_index().plot(title="Average Hourly Traffic (incl. weekdays & weekends)",\
                                        ax=axes[0], color="green", xticks=range(0,24,1))

# plot conversion rate per country
df_weekday.hour.value_counts().sort_index().div(len(df_weekday)).plot(
                                    title="Comparison of Weekday vs. Weekend traffic",\
                                        ax=axes[1], color="blue", xticks=range(0,24,1))
df_weekend.hour.value_counts().sort_index().div(len(df_weekend)).plot(ax=axes[1], color="red")

# create an identical legend for each plot
axes[0].legend(loc=2, handles=[average_day_leg])
axes[1].legend(loc=2, handles=[weekday_leg, weekend_leg])

# set y-axis labels
axes[0].set_xlabel("24Hour")
axes[1].set_xlabel("24Hour")
axes[0].set_ylabel("Total Hits")
axes[1].set_ylabel("Normalised Hits")
```

#### Looking at the above:
- The top graph shows the average traffic accross all days. We see a profile that seems consistent with the average commuter workday. 
- There is a ramp-up of activity during the early morning until about 08.00, after which traffic drops before climbing again in the afternoon and dropping sharply after 20.00 at night.
- Now looking at bottom graph. A broadly similar profile is seen between weekday and weekend data, however somewhat counter-intuitively we see that the weekend traffic spikes dramatically at 17.00, which would be more expected from the weekday (workplace) traffic.
- Further investigations would aggregrate more data from a number of weeks.

### 4.1 Referral Breakdown
- Let's examine a breakdown of the referral groups for the site
    - facebook
    - newspaper - internal homepage, section page or article
    - google
    - twitter
    - other

```python
def referrer_plot(df):
    """Function to create pie-chart of referral breakdown"""

    referrer_dict = {"facebook": 0, "internal_article": 0, "internal_homepage_or_section": 0, "google": 0, \
                 "twitter": 0, "other": 0}

    for index, value in df["page_referrer_group"].value_counts().iteritems():
        if "Facebook" in index:
            referrer_dict["facebook"] += value
        elif "INTERNAL: Homepage" in index:
            referrer_dict["internal_homepage_or_section"] += value
        elif "INTERNAL: Article" in index:
            referrer_dict["internal_article"] += value
        elif "INTERNAL: Section Fronts: " in index:
            referrer_dict["internal_homepage_or_section"] += value
        elif "Google" in index:
            referrer_dict["google"] += value
        elif "Twitter" in index:
            referrer_dict["twitter"] += value
        else:
            referrer_dict["other"] += value
        
    # set plot values
    labels = referrer_dict.keys()
    sizes = referrer_dict.values()

    # Plot a pie chart
    plt.pie(sizes, labels=labels, autopct='%d%%', shadow=True, startangle=0)   
    plt.axis('equal')
    plt.tight_layout()
    plt.title("Referral Breakdown", y=1.1)
    
    plt.show()
```

```python
referrer_plot(df)
```

###  5.1 Compare "Single-Click" vs. "Multi-Click" users
- See if there are obvious differences

```python
# create a dataframe that contains the total number of clicks during a session
session_counts_df = pd.DataFrame({'total_session_clicks' : df.groupby( [ "sessionId"])\
                               .size()}).reset_index().sort_values(by="total_session_clicks", ascending=False)

# Merge our original dataframe with the session counts dataframe
df = pd.merge(left=df, right=session_counts_df, how="left", \
                    left_on="sessionId", right_on="sessionId")

# if the total session length = 1; then we have a single click session

# first create the column (which will be binary) and set it to zero
df["single_click_session"] = 0

# now set the column "single_click_session" equal to 1 when we have a session_id length equal to 1
df.loc[(df['total_session_clicks'] == 1) , ['single_click_session']] = 1
```

```python
multi = df[df["single_click_session"]==0].copy()
single = df[df["single_click_session"]==1].copy()

multi_plot = df[df["single_click_session"]==0].hour.value_counts().sort_index()
single_plot = df[df["single_click_session"]==1].hour.value_counts().sort_index()

print "Percentage Single Click: %.2f%%" %(100*float(len(single))/len(df))
print "Percentage Multi Click: %.2f%%" %(100*float(len(multi))/len(df))

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(multi_plot, label='multi')
ax.plot(single_plot, label='single')

ax.set_xlabel('24hr')
ax.set_ylabel('Hits')
ax.set_title("Profile for traffic of single click vs. multi-click sessions")
ax.set_xticks(range(0,24,1)) ## ensuring we have appropriate x-ticks
ax.set_xlim(0,23)

# Now add the legend with some customizations.
legend = ax.legend(loc='best', shadow=True)
```

#### Looking at the above:
- The majority of users click multiple pages
- There is no obvious difference in the time-habits of single vs. multi-click users

### 6.3 Active vs. Dormant - Section Name Breakdown
- Compare the sections visited by the different users

```python
# Create a function to plot the largest components within the various groups

def group_plot(df, group):
    '''This function will split up our dataframe into groups
    and then plot the 10 biggest contributers to those groups'''
    
    ## Use groupby to split the data and then use .count on any column to counnt the 
    ## size of the subgroups - which contribute to the group we split upon
    sub_group_count = df.groupby(group)["date"].count()
      
    ## Now sort the values and take the top ten entries, then plot
    colors = ['y', 'b','r','g', 'magenta','#624ea7']
    sub_group_count.sort_values(ascending=False).head(6).plot(kind="barh",title=group, color=colors)
    
    plt.show()
    print "\n\n"
    
    return
```

```python
for user_group in user_group_list:
    print user_group.name
    group_plot(user_group, "sectionName")
```

#### Looking at the above:
- There is a very similar section breakdown between active and dormant users
- Frequent and infrequent users are both largely visiting the tv&showbiz and news sections of the site
